# Differential with python

Import matplotlib for plotting graph.  
Import numpy making N-dimensional array.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## variable
variable x is **domain(range of x data)** of the function.  
variable x1 is a **sample point.**  
variable tangent_range is linear domain df tangeut.(선을빼곡히넣어 매끄러운라인만듬=linspace)

```python
x = np.arange(-5, 5, 0.1) #x axis domain // domain = xaxis / range = y axis
x1 = 3
tangent_range = np.linspace(x1-2, x1+2,10)
```

```python
def f(x):
  return x**2

def slope(x):
  return 2*x

def tangent_line(x, x1, y1):
  return slope(x1)*(x-x1) + y1
```

```python
plt.figure()
plt.plot(x,f(x))
plt.scatter(x1,f(x1), color='red', s=20)
plt.plot(tangent_range, tangent_line(tangent_range, x1, f(x1)), color= 'red' , linewidth=2)
# plt.savefig('image/diff_img.png')
plt.show()
```
